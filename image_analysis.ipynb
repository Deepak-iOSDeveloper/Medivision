## let's build the cnn for each images

```python
cnn_brain_cancer_tumor = tf.keras.models.Sequential()
cnn_brain_tumor_not = tf.keras.models.Sequential()
cnn_brain_tumor_type = tf.keras.models.Sequential()
cnn_breast_tumor_type = tf.keras.models.Sequential()
cnn_heart_desease = tf.keras.models.Sequential()
cnn_brain_cancer_2 = tf.keras.models.Sequential()
cnn_breast_cancer_2 = tf.keras.models.Sequential()
cnn_cervical_cancer = tf.keras.models.Sequential()
cnn_kidney_cancer = tf.keras.models.Sequential()
cnn_lung_cancer = tf.keras.models.Sequential()
cnn_lymphoma_cancer = tf.keras.models.Sequential()
cnn_oral_cancer = tf.keras.models.Sequential()
cnn_pneumonia_tuborculosis = tf.keras.models.Sequential()
```

```python
cnn_brain_cancer_tumor.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[64, 64, 3]))
cnn_brain_tumor_not.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[64, 64, 3]))
cnn_brain_tumor_type.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[64, 64, 3]))
cnn_breast_tumor_type.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[64, 64, 3]))
cnn_heart_desease.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[64, 64, 3]))
cnn_brain_cancer_2.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[64, 64, 3]))
cnn_breast_cancer_2.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[64, 64, 3]))
cnn_cervical_cancer.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[64, 64, 3]))
cnn_kidney_cancer.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[64, 64, 3]))
cnn_lung_cancer.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[64, 64, 3]))
cnn_lymphoma_cancer.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[64, 64, 3]))
cnn_oral_cancer.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[64, 64, 3]))
cnn_pneumonia_tuborculosis.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[64, 64, 3]))
```

```python
cnn_brain_cancer_tumor.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_brain_tumor_not.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_brain_tumor_type.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_breast_tumor_type.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_heart_desease.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_brain_cancer_2.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_breast_cancer_2.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_cervical_cancer.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_kidney_cancer.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_lung_cancer.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_lymphoma_cancer.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_oral_cancer.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_pneumonia_tuborculosis.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn_brain_cancer_tumor.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
cnn_brain_tumor_not.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
cnn_brain_tumor_type.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
cnn_breast_tumor_type.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
cnn_heart_desease.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
cnn_brain_cancer_2.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
cnn_breast_cancer_2.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
cnn_cervical_cancer.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
cnn_kidney_cancer.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
cnn_lung_cancer.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
cnn_lymphoma_cancer.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
cnn_oral_cancer.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
cnn_pneumonia_tuborculosis.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
```

```python
cnn_brain_cancer_tumor.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_brain_tumor_not.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_brain_tumor_type.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_breast_tumor_type.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_heart_desease.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_brain_cancer_2.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_breast_cancer_2.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_cervical_cancer.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_kidney_cancer.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_lung_cancer.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_lymphoma_cancer.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_oral_cancer.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn_pneumonia_tuborculosis.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn_brain_cancer_tumor.add(tf.keras.layers.Flatten())
cnn_brain_tumor_not.add(tf.keras.layers.Flatten())
cnn_brain_tumor_type.add(tf.keras.layers.Flatten())
cnn_breast_tumor_type.add(tf.keras.layers.Flatten())
cnn_heart_desease.add(tf.keras.layers.Flatten())
cnn_brain_cancer_2.add(tf.keras.layers.Flatten())
cnn_breast_cancer_2.add(tf.keras.layers.Flatten())
cnn_cervical_cancer.add(tf.keras.layers.Flatten())
cnn_kidney_cancer.add(tf.keras.layers.Flatten())
cnn_lung_cancer.add(tf.keras.layers.Flatten())
cnn_lymphoma_cancer.add(tf.keras.layers.Flatten())
cnn_oral_cancer.add(tf.keras.layers.Flatten())
cnn_pneumonia_tuborculosis.add(tf.keras.layers.Flatten())
```

### now it's time to train out ANN

```python
cnn_brain_cancer_tumor.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_brain_tumor_not.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_brain_tumor_type.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_breast_tumor_type.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_heart_desease.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_brain_cancer_2.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_breast_cancer_2.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_cervical_cancer.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_kidney_cancer.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_lung_cancer.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_lymphoma_cancer.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_oral_cancer.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_pneumonia_tuborculosis.add(tf.keras.layers.Dense(units=128, activation="relu"))
```

```python
cnn_brain_cancer_tumor.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_brain_tumor_not.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_brain_tumor_type.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_breast_tumor_type.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_heart_desease.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_brain_cancer_2.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_breast_cancer_2.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_cervical_cancer.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_kidney_cancer.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_lung_cancer.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_lymphoma_cancer.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_oral_cancer.add(tf.keras.layers.Dense(units=128, activation="relu"))
cnn_pneumonia_tuborculosis.add(tf.keras.layers.Dense(units=128, activation="relu"))
```

```python
cnn_brain_cancer_tumor.add(tf.keras.layers.Dense(units=3, activation="softmax"))
cnn_brain_tumor_not.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
cnn_brain_tumor_type.add(tf.keras.layers.Dense(units=4, activation="softmax"))
cnn_breast_tumor_type.add(tf.keras.layers.Dense(units=3, activation="softmax"))
cnn_heart_desease.add(tf.keras.layers.Dense(units=4, activation="softmax"))
cnn_brain_cancer_2.add(tf.keras.layers.Dense(units=3, activation="softmax"))
cnn_breast_cancer_2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
cnn_cervical_cancer.add(tf.keras.layers.Dense(units=5, activation="softmax"))
cnn_kidney_cancer.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
cnn_lung_cancer.add(tf.keras.layers.Dense(units=5, activation="softmax"))
cnn_lymphoma_cancer.add(tf.keras.layers.Dense(units=3, activation="softmax"))
cnn_oral_cancer.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
cnn_pneumonia_tuborculosis.add(tf.keras.layers.Dense(units=3, activation="softmax"))
```

```python
cnn_brain_cancer_tumor.compile(optimizer="adam", loss = "categorical_crossentropy", metrics=["accuracy"])
cnn_brain_tumor_not.compile(optimizer="adam", loss = "binary_crossentropy", metrics=["accuracy"])
cnn_brain_tumor_type.compile(optimizer="adam", loss = "categorical_crossentropy", metrics=["accuracy"])
cnn_breast_tumor_type.compile(optimizer="adam", loss = "categorical_crossentropy", metrics=["accuracy"])
cnn_heart_desease.compile(optimizer="adam", loss = "categorical_crossentropy", metrics=["accuracy"])
cnn_brain_cancer_2.compile(optimizer="adam", loss = "categorical_crossentropy", metrics=["accuracy"])
cnn_breast_cancer_2.compile(optimizer="adam", loss = "binary_crossentropy", metrics=["accuracy"])
cnn_cervical_cancer.compile(optimizer="adam", loss = "categorical_crossentropy", metrics=["accuracy"])
cnn_kidney_cancer.compile(optimizer="adam", loss = "binary_crossentropy", metrics=["accuracy"])
cnn_lung_cancer.compile(optimizer="adam", loss = "categorical_crossentropy", metrics=["accuracy"])
cnn_lymphoma_cancer.compile(optimizer="adam", loss = "categorical_crossentropy", metrics=["accuracy"])
cnn_oral_cancer.compile(optimizer="adam", loss = "binary_crossentropy", metrics=["accuracy"])
cnn_pneumonia_tuborculosis.compile(optimizer="adam", loss = "categorical_crossentropy", metrics=["accuracy"])
```

```python
cnn_brain_cancer_tumor.fit(x = training_set_brain_cancer, validation_data= test_set_brain_cancer, epochs=35)
cnn_brain_tumor_not.fit(training_set_brain_tumor_or_not, validation_data= test_set_brain_tumor_or_not, epochs=35)
cnn_brain_tumor_type.fit(training_set_brain_tumor_type, validation_data= test_set_brain_tumor_type, epochs=35)
cnn_breast_tumor_type.fit(training_set_breast_tumor_type, validation_data= test_set_breast_tumor_type, epochs=35)
cnn_heart_desease.fit(training_set_heart_desease, validation_data= test_set_heart_desease, epochs=35)
cnn_brain_cancer_2.fit(training_set_brain_cancer_2, validation_data= test_set_brain_cancer_2, epochs=35)
cnn_breast_cancer_2.fit(training_set_breast_cancer_2, validation_data= test_set_breast_cancer_2, epochs=35)
cnn_cervical_cancer.fit(training_set_cervical_cancer, validation_data= test_set_cervical_cancer, epochs=35)
cnn_kidney_cancer.fit(training_set_kidney_cancer, validation_data= test_set_kidney_cancer, epochs=35)
cnn_lung_cancer.fit(training_set_lung_cancer, validation_data= test_set_lung_cancer, epochs=35)
cnn_lymphoma_cancer.fit(training_set_lymphoma, validation_data= test_set_lymphoma, epochs=35)
cnn_oral_cancer.fit(training_set_oral_cancer, validation_data= test_set_oral_cancer, epochs=35)
cnn_pneumonia_tuborculosis.fit(training_set_pneumonia_or_tuborculosis, validation_data= test_set_pneumonia_or_tuborculosis, epochs=35)
```
